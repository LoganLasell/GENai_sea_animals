<h3>
    The code below iterates through the full sea animal folder and stores all the images of our selected species "Clams" into a csv file, with their matching label, dimensions, file path, and image name.
</h3>

```python
import glob
import os
from pathlib import Path
from PIL import Image
import pandas as pd


# SETTINGS

parent_folder = "<folder with images>"
output_folder = "<desired_path>/sea_animals_standardized_gray"
csv_output_path = "sea_animals_dataset_gray.csv"
target_size = (200, 200)

target_species = "Clams"  # ← CHANGE THIS to your target species/folder name


#  IMAGE / FOLDER SETUP

os.makedirs(output_folder, exist_ok=True)
image_extensions = ('*.png', '*.jpg', '*.jpeg', '*.bmp', '*.gif')
records = []


# MAIN LOOP

for ext in image_extensions:
    for img_path in glob.glob(f"{parent_folder}/**/{ext}", recursive=True):
        img_path = Path(img_path)
        label = img_path.parent.name

        # Only process if it matches the target species
        if label.lower() != target_species.lower():
            continue

        img_name = img_path.name

        # Set new output path
        label_output_folder = Path(output_folder) / label
        label_output_folder.mkdir(parents=True, exist_ok=True)
        new_img_path = label_output_folder / img_name

        try:
            with Image.open(img_path) as img:
                img = img.convert('L')
                img = img.resize(target_size, Image.LANCZOS)
                img.save(new_img_path)
                width, height = img.size
        except Exception as e:
            print(f"Error processing {img_path}: {e}")
            continue

        records.append({
            "width": width,
            "height": height,
            "full_path": str(new_img_path.resolve()),
            "label": label,
            "image_name": img_name
        })


# SAVE CSV

df = pd.DataFrame(records)
df.to_csv(csv_output_path, index=False)

print(f"Processed {len(df)} images for species '{target_species}'.")
print(f"CSV saved to {csv_output_path}")
```

<h3>
    The code below uses the previously made csv file and transforms the csv into a Pytorch tensor file, which is just a different format of storage that is used in deep learning model to help with its efficiency. 
</h3>

```python
import torch
from torchvision import transforms
from PIL import Image
import pandas as pd


# SETTINGS

csv_path = "sea_animals_dataset_gray.csv"
image_size = (128, 128)  # Change this if needed
output_tensor_path = "sea_animals_tensor.pt"


# TRANSFORM

transform = transforms.Compose([
    transforms.Resize(image_size),
    transforms.ToTensor(),  # Converts to [0,1]
    transforms.Normalize([0.5], [0.5])  # Normalize grayscale to [-1, 1]
])


# LOAD IMAGES

df = pd.read_csv(csv_path)
tensor_list = []

for i, row in df.iterrows():
    try:
        img = Image.open(row['full_path']).convert('L')  # Grayscale
        img = transform(img)
        tensor_list.append(img)
    except Exception as e:
        print(f"Failed to load {row['full_path']}: {e}")


# STACK AND SAVE

all_images_tensor = torch.stack(tensor_list)
print(f"Tensor shape: {all_images_tensor.shape}")  # e.g., (N, 1, 128, 128)

torch.save(all_images_tensor, output_tensor_path)
print(f"Saved tensor dataset to {output_tensor_path}")
```

<h3>
    Loading dataset into the model. Ensuring that the images are all sized to a standard of 128 x 128 for accuracy with the images that the model will create.
</h3>

```python
import torch
#!pip install torchvision
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms


# DATASET CLASS

class SeaAnimalDataset(Dataset):
    def __init__(self, csv_path, transform=None):
        self.data = pd.read_csv(csv_path)
        self.transform = transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_path = self.data.iloc[idx]['full_path']
        img = Image.open(img_path).convert('L')  # Grayscale
        if self.transform:
            img = self.transform(img)
        return img



# TRANSFORMS

transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),  # [0,1]
    transforms.Normalize([0.5], [0.5])  # normalize grayscale images to [-1, 1]
])


# DATASET AND LOADER

images = torch.load("sea_animals_tensor.pt")
dataloader = DataLoader(images, batch_size=64, shuffle=True, num_workers=4)

```

<h3>
    Generator Creation: Part of GAN that creates fake images to test against real images.
</h3>

<h3>
    Each image starts with a random latent vector based off a real image. I then use a transposed convolution method to upscale each image to 1 channel, 128 x 128 product.  
</h3>

```python
import torch
import torch.nn as nn

class Generator(nn.Module):
    def __init__(self, latent_dim=100):
        super(Generator, self).__init__()
        self.latent_dim = latent_dim

        self.net = nn.Sequential(
            nn.ConvTranspose2d(latent_dim, 512, 4, 1, 0),  # (512, 4, 4)
            nn.BatchNorm2d(512),
            nn.ReLU(True),

            nn.ConvTranspose2d(512, 256, 4, 2, 1),  # (256, 8, 8)
            nn.BatchNorm2d(256),
            nn.ReLU(True),

            nn.ConvTranspose2d(256, 128, 4, 2, 1),  # (128, 16, 16)
            nn.BatchNorm2d(128),
            nn.ReLU(True),

            nn.ConvTranspose2d(128, 64, 4, 2, 1),   # (64, 32, 32)
            nn.BatchNorm2d(64),
            nn.ReLU(True),

            nn.ConvTranspose2d(64, 32, 4, 2, 1),    # (32, 64, 64)
            nn.BatchNorm2d(32),
            nn.ReLU(True),

            nn.ConvTranspose2d(32, 16, 4, 2, 1),    # (16, 128, 128)
            nn.BatchNorm2d(16),
            nn.ReLU(True),

            nn.ConvTranspose2d(16, 1, 3, 1, 1),     # (1, 128, 128)
            nn.Tanh()
        )

    def forward(self, z):
        z = z.view(z.size(0), self.latent_dim, 1, 1)
        return self.net(z)
```

<h3>
    Discriminator Creation: This is the part of the GAN that tries to detect fake images created by the Generator.
</h3>

<h3>
    The discriminator takes in the generator's fake image, sends it through convolutional layers meant to reduce the image down to just its spatial features which allow for it to extract only the needed details of the image to determine its validity.  
</h3

```python
class Discriminator(nn.Module):
    def __init__(self, image_size=(128, 128)):
        super(Discriminator, self).__init__()

        self.features = nn.Sequential(
            nn.Conv2d(1, 32, 4, 2, 1),   # (32, 100, 100)
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(32, 64, 4, 2, 1),  # (64, 50, 50)
            nn.BatchNorm2d(64),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(64, 128, 4, 2, 1), # (128, 25, 25)
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(128, 256, 4, 2, 1), # (256, 13, 13)
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.2, inplace=True),
        )

        # Dynamically determine the output size
        with torch.no_grad():
            dummy_input = torch.zeros(1, 1, *image_size)
            out = self.features(dummy_input)
            flatten_dim = out.view(1, -1).shape[1]

            print(f"Flatten dimension calculated: {flatten_dim}")

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(flatten_dim, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = self.features(x)
        return self.classifier(x)
```

<h3>
    Training the actual GAN model 
</h3>

<h4>
    Torch.optim is used in training the model, because it allows for optimization of the model's algorithm during actual training.
    You can also set the number of epochs you want the model to cycle through during training, the higher the number, the more realistic the model's images will be. 
</h4>

```python
import torch.optim as optim

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

generator = Generator().to(device)
discriminator = Discriminator().to(device)

criterion = nn.BCELoss()
optimizer_G = optim.Adam(generator.parameters(), lr=2e-4, betas=(0.5, 0.999))
optimizer_D = optim.Adam(discriminator.parameters(), lr=2e-4, betas=(0.5, 0.999))

noise_dim = 100
epochs = 200

for epoch in range(epochs):
    for imgs in dataloader:
        imgs = imgs.to(device)

        # Train Discriminator
        real_labels = torch.ones(imgs.size(0), 1).to(device)
        fake_labels = torch.zeros(imgs.size(0), 1).to(device)

        outputs = discriminator(imgs)
        d_loss_real = criterion(outputs, real_labels)

        z = torch.randn(imgs.size(0), noise_dim).to(device)
        fake_imgs = generator(z)
        outputs = discriminator(fake_imgs.detach())
        d_loss_fake = criterion(outputs, fake_labels)

        d_loss = d_loss_real + d_loss_fake

        optimizer_D.zero_grad()
        d_loss.backward()
        optimizer_D.step()

        # Train Generator
        z = torch.randn(imgs.size(0), noise_dim).to(device)
        fake_imgs = generator(z)
        outputs = discriminator(fake_imgs)
        g_loss = criterion(outputs, real_labels)

        optimizer_G.zero_grad()
        g_loss.backward()
        optimizer_G.step()

    print(f"Epoch [{epoch+1}/{epochs}]  Loss_D: {d_loss.item():.4f}, Loss_G: {g_loss.item():.4f}")
```

<h3>
    To show the fake images created at the end of training, I use torchvision.utils to create a grid of fake images and use matplotlib to plot and display those images. 
</h3>

```python
import torch
import torchvision.utils as vutils
import matplotlib.pyplot as plt

# Generate fake images from your generator
with torch.no_grad():
    fake = generator(z).detach().cpu()  # shape: (batch_size, channels, H, W)

# Create a grid of images
grid = vutils.make_grid(fake, nrow=8, normalize=True, value_range=(-1, 1))  # adjust nrow if needed

# Convert to NumPy for matplotlib
np_grid = grid.permute(1, 2, 0).numpy()  # shape: (H, W, C)

# Plot
plt.figure(figsize=(10, 10))
if np_grid.shape[2] == 1:
    plt.imshow(np_grid[:, :, 0], cmap="gray")
else:
    plt.imshow(np_grid)
plt.axis("off")
plt.title(f"Generated Samples at Epoch {epoch+1}")
plt.show()
```
